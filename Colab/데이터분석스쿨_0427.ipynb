## 변수의 속성

* 영문과 숫자를 사용할 수 있지만, 숫자로 시작하지는 못합니다.
* 특수문자는 사용하지 않아요.(언더바(_): 스네이크 표기법은 가능)
* 예약어는 사용하지 않습니다.(if, elif, while, for, etc)
* 대소문자는 구분합니다.
* 언더바로만 사용하거나 언더바로 시작할 수 있습니다.
* 대문자로 시작하는 변수를 사용할 수 있지만, 관습적으로 대문자로 시작하는 변수는 클래스(Class)로 만들기 때문에 대체로 소문자를 사용합니다.
* Class는 보통 파스칼 표기법을 따릅니다. 다만 회사 컨벤션마다 다릅니다.

## 논리연산

```python
# True : 1
# False : 0

# and : 논리곱
# or : 논리합
# not : 부정
```

```python
# True and False
# 둘 중 하나라도 False 라면 False
# and는 위와 같은 속성 때문에 둘 다 만족해야 하는 '조건'에 사용

for i in range(101):
  if i % 3 == 0 and i % 5 == 0:
      print(i)
```

```python
# True or False
# 둘 중 하나라도 True 라면 True
# and는 위와 같은 속성 때문에 둘 다 만족해야 하는 '조건'에 사용

for i in range(101):
  if i % 3 == 0 or i % 5 == 0:
      print(i)
```

```python
for i in range(101):
    if (i % 3 == 0 and i % 5 == 0) or i % 2 == 0:   # and가 항상 or보다 먼저 계산됨!
        print(i)
```

```python
for i in range(101):
    if i % 3 == 0 or (i % 5 == 0 and i % 2 == 0):
        print(i)
```

```python
a = 10 and 20 # and는 앞에 값에 따라서 뒤의 값도 보아야 함. 따라서 20이 출력됨
print(a)
```

```python
a = '' and 10 # 앞의 값이 False 이기 때문에, 뒤의 값은 보지 않음
print(a)
```

```python
a = 10 or 20 # or는 앞의 값이 차있으면 뒤에 값은 볼 필요가 없음
print(a)
```

```python
a = '' or 10  # 앞의 값이 False 이기 때문에, 뒤의 값을 보게 됨
print(a)
```

```python
userName = '훈'
userName = userName or '훈' # userNmae 이 이미 True가 나왔기 때문에 뒤에 '훈'을 볼 필요가 없음. 파이썬도 역시 뒤에 볼 필요가 없는 경우엔 뒤의 코드를 실행하지 않음
print(userName)
```

## 그리디 알고리즘

* 욕심쟁이 알고리즘
* 가장 최선을 선택한 것이 최종적으로도 가장 최선의 선택임

```python
# (연습문제) 10000원을 들고 가서 1200원짜리 아이스크림 먹었을 때 잔돈의 갯수가 가장 적은 경우를 구하라

잔돈 = 8800

오천원갯수 = 잔돈 // 5000
잔돈 = 잔돈 - 오천원갯수 * 5000

천원의갯수 = 잔돈 // 1000
잔돈 = 잔돈 - 천원의갯수 * 1000

오백원갯수 = 잔돈 // 500
잔돈 = 잔돈 - 오백원갯수 * 500

백원갯수 = 잔돈 // 100
잔돈 = 잔돈 - 백원갯수 * 100

print(f'오천원은 {오천원갯수}개, \
천원은 {천원의갯수}개, \
오백원은 {오백원갯수}개, \
백원 {백원갯수}개를 거슬러 드립니다.')
```
