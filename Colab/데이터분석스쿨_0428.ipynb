```python
# 데이터 분석에서 대부분의 재귀함수 코드는 for문으로 대체하는게 좋음! (재귀함수는 매우 비효율적이기 때문에 최후의 수단으로 사용)
# 다만, 메모라이징 기법을 사용하면 좋긴 함
# 다이나믹 프로그래밍 등 필수적으로 사용되는 곳이 있기도 함
# 팩토리얼 만들기, 문자 거꾸로 출력하기 정도 배우고 지나가는 것이 좋음
```

```python
# 팩토리얼 만들기

def factorial(n):
    if n == 1:
        return n
    return n * factorial(n-1)   # factorial로 자기 자신을 다시 호출

factorial(5)

#                 return 값
# factorial(5) == 5 * factorial(4)
# factorial(4) == 4 * factorial(3)
# factorial(3) == 3 * factorial(2)
# factorial(2) == 2 * factorial(1)  
# factorial(1) == 1
```

```python
# for문으로 치환하면...

result = 1
for i in range(1, 6):
    result *= i

result
```

```python
# 문자 거꾸로 출력하기

def string_reverse(s):
    if len(s) == 1:
        return s               # 코드가 매우 길면 가능하면 앞쪽에서 if로 빠르게 리턴해주는 것이 좋음(얼리리턴)
    return string_reverse(s[1:]) + s[0]     # string_reverse(s[1:])가 다시 위에서부터 반복함

string_reverse('hello')

# string_reverse('hello') == string_reverse('ello') + 'h'
# string_reverse('ello') == string_reverse('llo') + 'e'
# string_reverse('llo') == string_reverse('lo') + 'l'
# string_reverse('lo') == string_reverse('o') + 'l'
# string_reverse('o') == 'o'
```

```python
# for문으로 치환하면...

s = 'leehojun'
result = ''
for i in s:
    result = i + result

result
```

```python
# (연습문제3) 3의 n제곱을 구하는 함수를 만들어주세요. (재귀함수를 이용하여 만들어야하며, x는 0이 아닙니다) (n > 1)

def 제곱(n):
    if n == 0:
        return 1
    return 3 * 제곱(n-1)

제곱(3)
제곱(4)    

# 제곱(4)    3 * 제곱(3)    3 * (3 * 3 * 3)
# 제곱(3)    3 * 제곱(2)    3 * (3 * 3)
# 제곱(2)    3 * 제곱(1)    3 * 3
# 제곱(1)    3 * 제곱(0)    3 * 1
# 제곱(0)    1
```

```python
# (연습문제3) x의 n제곱을 구하는 함수를 만들어주세요. (재귀함수를 이용하여 만들어야하며, x는 0이 아닙니다) (x, n > 1)

def 제곱(x, n):
    if n == 0:
        return 1
    return x * 제곱(x, n-1)

제곱(3,4)

# 제곱(3, 4)    3 * 제곱(3, 3)    3 * (3 * 3 * 3)
# 제곱(3, 3)    3 * 제곱(3, 2)    3 * (3 * 3)
# 제곱(3, 2)    3 * 제곱(3, 1)    3 * 3
# 제곱(3, 1)    3 * 제곱(3, 0)    3 * 1
# 제곱(3, 0)    1
```

```python
# 연습문제

l = [[1, 10, 'leehojun'], 
     [20, 30, 'hojun'], 
     [10, 20, 'weniv'], 
     [1, 2, 'likelion'], 
     [55, 11, 'sun']]

# 1. 글자 수 대로 정렬해주세요.
def 함수(x):
    return len(x[2])

print(list(sorted(l, key=함수)))
print(list(sorted(l, key=lambda x:len(x[2]))))

# 2. 맨 앞에 위치한 숫자대로 정렬해주세요.
def 함수2(x):
    return x[0]

print(list(sorted(l)))
print(list(sorted(l, key=함수2)))
print(list(sorted(l, key=lambda x:x[0])))

# 3. 중앙에 위치한 값대로 정렬해주세요.
def 함수3(x):
    return x[1]

print(list(sorted(l)))
print(list(sorted(l, key=함수3)))
print(list(sorted(l, key=lambda x:x[1])))
```
